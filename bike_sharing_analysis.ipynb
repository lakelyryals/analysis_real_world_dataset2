```python
# ðŸ“˜ Bike Sharing Data Analysis

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("day.csv")

# Preprocessing
df["dteday"] = pd.to_datetime(df["dteday"])
df["season_label"] = df["season"].map({1: "Winter", 2: "Spring", 3: "Summer", 4: "Fall"})
df["weekday_label"] = df["weekday"].map({0: "Sun", 1: "Mon", 2: "Tue", 3: "Wed", 4: "Thu", 5: "Fri", 6: "Sat"})

# Summary statistics
df[["temp", "hum", "windspeed", "casual", "registered", "cnt"]].describe()
```

```python
# Daily Rentals over time
plt.figure(figsize=(12, 5))
sns.lineplot(x="dteday", y="cnt", data=df)
plt.title("Total Daily Bike Rentals Over Time")
plt.xlabel("Date")
plt.ylabel("Count of Rentals")
plt.tight_layout()
plt.show()
```

```python
#Rentals by Season

plt.figure(figsize=(8, 5))
sns.boxplot(x="season_label", y="cnt", hue="season_label", data=df, palette="pastel", legend=False)
plt.title("Bike Rentals by Season")
plt.xlabel("Season")
plt.ylabel("Rental Count")
plt.tight_layout()
plt.show()
```

```python
# Temperature vs rental counts

plt.figure(figsize=(8, 5))
sns.scatterplot(x="temp", y="cnt", data=df, alpha=0.6)
sns.regplot(x="temp", y="cnt", data=df, scatter=False, color="red", ci=None)
plt.title("Temperature vs. Total Rentals")
plt.xlabel("Normalized Temperature")
plt.ylabel("Rental Count")
plt.tight_layout()
plt.show()
```

```python
# Rentals by day of week

plt.figure(figsize=(8, 5))
sns.boxplot(x="weekday_label", y="cnt", hue="weekday_label", data=df,
            palette="pastel", order=["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"],
            legend=False)
plt.title("Bike Rentals by Weekday")
plt.xlabel("Day of the Week")
plt.ylabel("Rental Count")
plt.tight_layout()
plt.show()
```

```python
#Humidity vs rental count
plt.figure(figsize=(8, 5))
sns.scatterplot(x="hum", y="cnt", data=df, alpha=0.6)
sns.regplot(x="hum", y="cnt", data=df, scatter=False, color="red", ci=None)
plt.title("Humidity vs. Total Rentals")
plt.xlabel("Humidity")
plt.ylabel("Rental Count")
plt.tight_layout()
plt.show()
```

```python
# Casual vs registered users

plt.figure(figsize=(8, 5))
sns.scatterplot(x="casual", y="registered", data=df, alpha=0.6)
plt.title("Casual vs Registered Rentals")
plt.xlabel("Casual Users")
plt.ylabel("Registered Users")
plt.tight_layout()
plt.show()
```
